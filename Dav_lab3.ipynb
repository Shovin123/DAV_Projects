```python
import matplotlib.pyplot as plt
import numpy as np
xpoint=np.array([0,6])
ypoint=np.array([0,250])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
ypoint=np.array([3,8,1,10,5,7])
plt.plot(ypoint)
plt.show()
plt.plot(ypoint, marker='o')
plt.show()
plt.plot(ypoint, 'o:r')
plt.show()
plt.plot(ypoint, 'o--y')
plt.show()
plt.plot(ypoint, 'o-.b')
plt.show()
plt.plot(ypoint, marker='*', ms=20, mfc='r')
```
